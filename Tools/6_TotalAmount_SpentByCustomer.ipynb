# Task

```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf = conf)
```

```python
input = sc.textFile("/content/drive/MyDrive/SparkCourse/customer-orders.csv")
```

```python
def parseLine(line):
    fields = line.split(',')
    customerID = int(fields[0])
    amount = float(fields[2])
    return (customerID, amount)
```

```python
rdd = input.map(parseLine)
```

```python
totalsCustomer = rdd.mapValues(lambda x: (x, 1)).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1])).sortByKey()
```

```python
results = totalsCustomer.collect()
```

```python
for result in results:
    print(result)
```

```python
def extractCustomerPricePairs(line):
  fields = line.split(',')
  return (int(fields[0]), float(fields[2]))
```

```python
mappedInput = input.map(extractCustomerPricePairs)
totalByCustomer = mappedInput.reduceByKey(lambda x, y: x +y).sortByKey()
results = totalByCustomer.collect()
```

```python
for result in results:
  print(result)
```
